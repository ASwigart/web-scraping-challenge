```python
# import the dependency
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import requests
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db= client.mars_db
collection = db.items
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# url of page to be scraped
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
html=browser.html
```

```python
# retrieval 
response = requests.get(url)
# introduce beautiful soup and parse
soup= bs(html, 'html.parser')
slide_elem = soup.select_one('div.list_text')
```

```python
print(soup.prettify())
```

```python
# tester cell
results =slide_elem.find('div', class_='content_title').text
results
```

```python
results = soup.find_all('div', class_='col-md-8')
headlines = []
# loop
for result in results:
    title = result.find('div', class_='content_title').text.strip()
    preview= result.find('div', class_='article_teaser_body').text
    
   
    print('----------------------')
    print(title)
    print(preview)
    
    post = {
    'title': title,
    'preview': preview,
    }
#      collection.insert_one(post)

    headlines.append(post)
    try:
        browser.links.find_by_partial_text('MORE'). click()
    except:
        print("scrape complete")
```

```python
headlines
```

```python
for h in headlines:
    collection.insert_one(h) 
```
